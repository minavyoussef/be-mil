```python
import pandas as pd
import os
```

```python
DATA_DIR = './data'
RESOLUTION = 'daily'
REGION = 'us'
INSTRUMENT = './nyse stocks/2/unh.us.txt'
```

```python
data_file = os.path.normpath(os.path.join(DATA_DIR, RESOLUTION, REGION, INSTRUMENT))
data_file
```

```python
df = pd.read_csv(data_file)
df
```

```python
df.Date = pd.to_datetime(df.Date, format='%Y%m%d')
df
```

```python
df['Close_t1'] = df['Close'].shift(-1)
df['Close_delta'] = df['Close_t1'] - df['Close']
df = df[['Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'Close_t1', 'Close_delta']]
df = df[:-1]
df
```

```python
from sklearn.preprocessing import StandardScaler

df[['Open', 'High', 'Low', 'Close', 'Volume']] = \
    StandardScaler().fit_transform(df[['Open', 'High', 'Low', 'Close', 'Volume']])
```
